```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot data from CSV and save as PNG
def plot_csv_to_png(csv_file, output_folder):
    # Read the CSV data into a pandas DataFrame
    df = pd.read_csv(csv_file)
    
    # Extract columns for plotting (adjust if the column names are different)
    x = df['Step']
    y = df['Value']
    
    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, linestyle='-', color='b', label='Value')  # Only the line, no dots
    
    # Set labels and title
    plt.xlabel('Step')
    plt.ylabel('Value')
    plt.title(f'Graph for {os.path.basename(csv_file)}')
    plt.legend()
    
    # Make the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)
    
    # Save the plot as PNG
    output_file = os.path.join(output_folder, f"{os.path.splitext(os.path.basename(csv_file))[0]}.png")
    plt.savefig(output_file)
    plt.close()

# Main function to process all CSV files in a directory
def process_tensorboard_data(input_folder, output_folder):
    # Loop through all files in the input folder
    for file_name in os.listdir(input_folder):
        file_path = os.path.join(input_folder, file_name)
        
        # Check if the file is a CSV file
        if os.path.isfile(file_path) and file_name.endswith('.csv'):
            print(f"Processing {file_name}...")
            plot_csv_to_png(file_path, output_folder)
            print(f"Graph saved for {file_name} as PNG.")


# Set your directories here
input_folder = r'C:\Users\cyberwitch\Documents\portfolio\ALTR_PROJECT\data\tf_logs'  # Replace with the path to your TensorBoard CSV files
output_folder = r'C:\Users\cyberwitch\Documents\portfolio\ALTR_PROJECT\graphs'         # Replace with the path where you want to save the PNGs

# Process the data
process_tensorboard_data(input_folder, output_folder)
```

```python
import pandas as pd
import googlemaps
from datetime import datetime
import time
from decouple import config

# Your 
#  Maps API key
API_KEY = 'ADD ME'
gmaps = googlemaps.Client(key=API_KEY)

# Function to get the distance between two addresses
def get_distance(pickup_address, dropoff_address):
    try:
        directions = gmaps.directions(pickup_address, dropoff_address, mode="driving")
        if directions:
            distance = directions[0]['legs'][0]['distance']['value']  # distance in meters
            return distance / 1000  # Convert to kilometers
        else:
            return 0
    except Exception as e:
        print(f"Error getting distance: {e}")
        return 0

# Function to calculate idle time percentage per asset
def calculate_idle_time_percentage(pickup_time, appt_time):
    try:
        pickup_time_obj = datetime.strptime(pickup_time, "%H:%M:%S")
        appt_time_obj = datetime.strptime(appt_time, "%H:%M:%S")
        idle_time = (appt_time_obj - pickup_time_obj).seconds / 3600  # Idle time in hours
        return idle_time * 100 / 8  # Assuming 8-hour shifts
    except Exception as e:
        print(f"Error calculating idle time: {e}")
        return 0

# Function to calculate the average utilization rate per asset
def calculate_utilization_rate(utilized_time, total_shift_time):
    try:
        utilization_rate = (utilized_time / total_shift_time) * 100
        return utilization_rate
    except ZeroDivisionError:
        return 0

# Load CSV file
df = pd.read_csv(r"C:\Users\cyberwitch\Documents\portfolio\ALTR_PROJECT\data\2024-06-06.csv")

# Initialize dictionary to store results
results = {
    'asset': [],
    'average_distance': [],
    'idle_time_percentage': [],
    'average_utilization_rate': []
}

# For each unique asset, calculate the necessary metrics
for asset in df['asset'].unique():
    # Filter the data for the current asset
    asset_data = df[df['asset'] == asset]

    total_distance = 0
    total_utilized_time = 0
    total_idle_time = 0
    total_shifts = 0

    for index, row in asset_data.iterrows():
        # Calculate distance for the ride
        distance = get_distance(row['pickup_address'], row['dropoff_address'])
        total_distance += distance

        # Calculate idle time percentage
        idle_time_percent = calculate_idle_time_percentage(row['pickup_time'], row['appt_time'])
        total_idle_time += idle_time_percent

        # Calculate utilized time (assuming the utilized time is the duration of the ride in hours)
        pickup_time_obj = datetime.strptime(row['pickup_time'], "%H:%M:%S")
        appt_time_obj = datetime.strptime(row['appt_time'], "%H:%M:%S")
        ride_duration = (appt_time_obj - pickup_time_obj).seconds / 3600  # Duration in hours
        total_utilized_time += ride_duration
        total_shifts += 1

    # Calculate averages
    average_distance = total_distance / total_shifts if total_shifts > 0 else 0
    average_idle_time = total_idle_time / total_shifts if total_shifts > 0 else 0
    average_utilization_rate = calculate_utilization_rate(total_utilized_time, total_shifts * 8)  # Assuming 8-hour shifts

    # Store the results
    results['asset'].append(asset)
    results['average_distance'].append(average_distance)
    results['idle_time_percentage'].append(average_idle_time)
    results['average_utilization_rate'].append(average_utilization_rate)

    # Sleep to prevent exceeding API rate limits (if needed)
    time.sleep(1)

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Display the results
print(results_df)

# Save the results to a new CSV file
results_df.to_csv("asset_metrics.csv", index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
from PIL import Image
from matplotlib.animation import FuncAnimation

# Load map image (make sure you have it in the right path)
map_image_path = r"C:\Users\cyberwitch\Documents\portfolio\ALTR_PROJECT\animations\ride_clustering\sj_map.JPG"  # Your image file path here
map_image = Image.open(map_image_path)

# Define the coordinates for the 11 assets (central points)
# Replace the (x, y) tuples below with your actual pixel coordinates
assets = [
    (100, 150), (200, 250), (300, 350), (400, 450), (500, 550),
    (600, 650), (700, 750), (800, 850), (900, 950), (1000, 1050), (1100, 1150)
]

# Initial random data points (example - adjust as needed)
data_points = np.random.rand(100, 2) * np.array(map_image.size)

# Function to assign points to nearest assets (cluster)
def assign_to_clusters(points, assets):
    distances = np.array([[distance.euclidean(point, asset) for asset in assets] for point in points])
    return np.argmin(distances, axis=1)

# Function to gradually move points towards clusters
def move_points_toward_clusters(points, clusters, assets, move_fraction=0.05):
    for i, cluster in enumerate(clusters):
        cluster_points = points[cluster == i]
        if len(cluster_points) > 0:
            # Move the points gradually towards the asset's location (central point)
            move_vector = (assets[i] - cluster_points) * move_fraction
            points[cluster == i] += move_vector
    return points

# Set up the plot and animation
fig, ax = plt.subplots()
ax.imshow(map_image)

# Create a function for updating the animation frame
def update(frame):
    ax.clear()
    ax.imshow(map_image)
    
    # Assign points to the nearest asset
    clusters = assign_to_clusters(data_points, assets)
    
    # Gradually move points toward the clusters
    data_points[:] = move_points_toward_clusters(data_points, clusters, assets, move_fraction=0.05)
    
    # Plot the data points
    ax.scatter(data_points[:, 0], data_points[:, 1], c=clusters, cmap="viridis", s=10)
    
    # Plot the assets (central points of clusters)
    assets_x, assets_y = zip(*assets)
    ax.scatter(assets_x, assets_y, c='red', marker='x', s=100, label="Assets")
    
    # Set labels and title
    ax.set_title(f"Clustering Progress - Frame {frame}")
    ax.legend()

# Set up the animation
num_frames = 100  # Adjust based on how long you want the animation to last
ani = FuncAnimation(fig, update, frames=num_frames, repeat=False)

# Save the animation as a GIF
ani.save('clustering_process.gif', writer='imagemagick', fps=10)

plt.show()
```
