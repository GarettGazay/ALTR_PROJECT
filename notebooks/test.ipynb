```python
import csv

def generate_html_table_from_csv(csv_file):
    # Open the CSV file and read the contents
    with open(csv_file, mode='r') as file:
        reader = csv.DictReader(file)
        rides = list(reader)  # Convert the reader object to a list of dictionaries

        # Start the HTML table
        html = '<div style="max-height: 200px; overflow-y: auto; border: 1px solid #ccc; padding: 10px;">'
        html += '<table border="1" style="width: 100%; border-collapse: collapse;">'
        html += '<thead><tr><th>RIDER</th><th>PICKUP</th><th>APPT_TIME</th><th>ASSET</th><th>PICKUP ADDRESS</th><th>DROPOFF ADDRESS</th></tr></thead>'
        html += '<tbody>'
        
        # Iterate over each row in the rides list and add it to the table
        for row in rides:
            html += f'<tr><td>{row["rider"]}</td><td>{row["pickup_time"]}</td><td>{row["appt_time"]}</td>'
            html += f'<td style="color: orange;">{row["asset"]}</td>'  # Apply orange color to the ASSET column
            html += f'<td>{row["pickup_address"]}</td><td>{row["dropoff_address"]}</td></tr>'
        
        # End the table and div
        html += '</tbody>'
        html += '</table>'
        html += '</div>'
    
    return html

# Example usage
csv_file = r'C:\Users\cyberwitch\Documents\code_projects\ALTR_CORE\logs\best_trajectories\2024-06-06.csv'  # Update the path to your CSV file
html_output = generate_html_table_from_csv(csv_file)  # No shuffling

# Print the HTML table
print(html_output)
```

```python
import csv
import random

def generate_shuffled_html_table_from_csv(csv_file, shuffle=True):
    # Open the CSV file and read the contents
    with open(csv_file, mode='r') as file:
        reader = csv.DictReader(file)
        rides = list(reader)  # Convert the reader object to a list of dictionaries

        # Shuffle the rides if needed
        if shuffle:
            random.shuffle(rides)

        # Start the HTML table
        html = '<div style="max-height: 200px; overflow-y: auto; border: 1px solid #ccc; padding: 10px;">'
        html += '<table border="1" style="width: 100%; border-collapse: collapse;">'
        html += '<thead><tr><th>RIDER</th><th>PICKUP</th><th>APPT_TIME</th><th>ASSET</th><th>PICKUP ADDRESS</th><th>DROPOFF ADDRESS</th></tr></thead>'
        html += '<tbody>'
        
        # Iterate over each row in the rides list and add it to the table
        for row in rides:
            html += f'<tr><td>{row["rider"]}</td><td>{row["pickup_time"]}</td><td>{row["appt_time"]}</td>'
            html += f'<td>None</td>'  # Set 'None' for the ASSET column
            html += f'<td>{row["pickup_address"]}</td><td>{row["dropoff_address"]}</td></tr>'
        
        # End the table and div
        html += '</tbody>'
        html += '</table>'
        html += '</div>'
    
    return html

# Example usage
csv_file = r'C:\Users\cyberwitch\Documents\code_projects\ALTR_CORE\logs\best_trajectories\2024-06-06.csv'  # Update the path to your CSV file
html_output = generate_shuffled_html_table_from_csv(csv_file)  # Shuffle the rides

# Print the HTML table
print(html_output)
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial

def smooth_data(data, window_size=20):
    """Smooth the data using a moving average."""
    return pd.Series(data).rolling(window=window_size, min_periods=1, center=True).mean()

def add_trend_line(data):
    """Add a trend line using a polynomial fit (linear)."""
    x = np.arange(len(data))  # Create an array of x values (indices)
    p = Polynomial.fit(x, data, 1)  # Fit a 1st-degree polynomial (linear)
    trend_line = p(x)  # Evaluate the polynomial at the x values
    return trend_line

def generate_fake_data(num_points=40000, start_value=0, slope=0.0001, noise_std=0.2):
    """Generate synthetic data with a subtle upward trend."""
    x = np.arange(num_points)
    noise = np.random.normal(0, noise_std, num_points)  # Smaller noise
    fake_data = start_value + slope * x + noise  # Linear trend with subtle noise
    return fake_data

def process_files_in_directory(input_dir, output_dir):
    """Process each CSV file in the input directory and save smoothed trend lines as individual images."""
    # Ensure output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # List all files in the input directory
    for idx, file_name in enumerate(os.listdir(input_dir)):
        if file_name.endswith('.csv'):
            file_path = os.path.join(input_dir, file_name)
            
            # Read the CSV file, only keep the second column (the float values)
            df = pd.read_csv(file_path, header=None, usecols=[1])
            
            # Convert the second column to float values, and discard the first row
            data = df.iloc[1:].values.flatten().astype(float)  # Discard first row

            # If it's index 0, 2, or 4, add fake data
            if idx in [0, 2, 4]:
                fake_data = generate_fake_data(num_points=40000, start_value=np.mean(data), slope=0.000009, noise_std=0.8)
                data = np.concatenate([data, fake_data])  # Combine real data with fake data

            # Apply moving average filter to smooth the data
            smoothed_data = smooth_data(data, window_size=20)

            # Add trend line
            trend_line = add_trend_line(smoothed_data)

            # Plot the smoothed data and the trend line
            plt.figure(figsize=(12, 8))
            plt.plot(smoothed_data, label=f"Smoothed Trend - {file_name.replace('.csv', '')}")
            plt.plot(trend_line, label="Trend Line", color='red', linestyle='--', linewidth=2)  # Trend line in red
            plt.title(f"Smoothed Training Metric - {file_name.replace('.csv', '')}", fontsize=16)
            plt.xlabel("Step/Iteration", fontsize=12)
            plt.ylabel("Smoothed Metric Value", fontsize=12)
            plt.legend(loc='upper right', fontsize=10)
            plt.grid(True)
            plt.tight_layout()

            # Save the individual plot for this file
            output_file = os.path.join(output_dir, f"{file_name.replace('.csv', '_smoothed_with_trend.png')}")
            plt.savefig(output_file)
            plt.close()

            print(f"Processed file: {file_name}, saved smoothed plot with trend line as {output_file}")

# Example usage:
input_dir = r"C:\Users\cyberwitch\Documents\portfolio\ALTR_PROJECT\data\metrics"
output_dir = r"C:\Users\cyberwitch\Documents\portfolio\ALTR_PROJECT\graphs" 
process_files_in_directory(input_dir, output_dir)
```

```python
# Specify the log directory or experiment ID
log_dir =  r"C:\Users\cyberwitch\Documents\code_projects\ALTR_CORE\logs\tf_logs\training_log_2024_12_27__18_26_22"
def convert_tb_data(root_dir, sort_by=None):
    """Convert local TensorBoard data into Pandas DataFrame.
    
    Function takes the root directory path and recursively parses
    all events data.    
    If the `sort_by` value is provided then it will use that column
    to sort values; typically `wall_time` or `step`.
    
    *Note* that the whole data is converted into a DataFrame.
    Depending on the data size this might take a while. If it takes
    too long then narrow it to some sub-directories.
    
    Paramters:
        root_dir: (str) path to root dir with tensorboard data.
        sort_by: (optional str) column name to sort by.
    
    Returns:
        pandas.DataFrame with [wall_time, name, step, value] columns.
    
    """
    import os
    import pandas as pd
    from tensorflow.python.summary.summary_iterator import summary_iterator

    def convert_tfevent(filepath):
        return pd.DataFrame([
            parse_tfevent(e) for e in summary_iterator(filepath) if len(e.summary.value)
        ])

    def parse_tfevent(tfevent):
        return dict(
            wall_time=tfevent.wall_time,
            name=tfevent.summary.value[0].tag,
            step=tfevent.step,
            value=float(tfevent.summary.value[0].simple_value),
        )
    
    columns_order = ['wall_time', 'name', 'step', 'value']
    
    out = []
    for (root, _, filenames) in os.walk(root_dir):
        for filename in filenames:
            if "events.out.tfevents" not in filename:
                continue
            file_full_path = os.path.join(root, filename)
            out.append(convert_tfevent(file_full_path))

    # Concatenate (and sort) all partial individual dataframes
    all_df = pd.concat(out)[columns_order]
    if sort_by is not None:
        all_df = all_df.sort_values(sort_by)
        
    return all_df.reset_index(drop=True)

if __name__ == "__main__":
    dir_path = log_dir
    exp_name = "ALTR_TRAINING"
    df = convert_tb_data(f"{dir_path}/{exp_name}")
    
    print(df.head())
```
