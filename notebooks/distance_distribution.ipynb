```python
import pandas as pd
import googlemaps
import folium
from folium.plugins import HeatMap
from folium import FeatureGroup

# Initialize the Google Maps API client with your API key
gmaps = googlemaps.Client(key='AIzaSyAYtE2r26Tlw_4tl_BHkpGzj3Kt2Jy5oBo')

# Load your CSV file
df = pd.read_csv(r'C:\Users\cyberwitch\Documents\portfolio\ALTR_PROJECT\data\training_data_2024-06-03_ca6c8549738ba3b0.csv')  # Replace with the path to your CSV file

# Function to geocode addresses (convert address to lat/lng)
def geocode_address(address):
    geocode_result = gmaps.geocode(address)
    if geocode_result:
        return geocode_result[0]['geometry']['location']
    return None

# Initialize the map (centered around an approximate location)
map_center = [37.423021, -122.083739]  # Default to San Francisco
mymap = folium.Map(location=map_center, zoom_start=13)

# Create a list to store latitude and longitude of pickup and dropoff points for the heatmap
heatmap_data = []

# Create FeatureGroups for both markers and the heatmap
marker_group = FeatureGroup(name='Markers')
heatmap_group = FeatureGroup(name='Heatmap')

# Plot pickup and dropoff locations
for idx, row in df.iterrows():
    pickup_location = geocode_address(row['pickup_address'])
    dropoff_location = geocode_address(row['dropoff_address'])

    if pickup_location:
        folium.Marker(
            location=[pickup_location['lat'], pickup_location['lng']],
            popup=f"Pickup: {row['pickup_address']}",
            icon=folium.Icon(color='blue')
        ).add_to(marker_group)
        # Add the pickup location to the heatmap data
        heatmap_data.append([pickup_location['lat'], pickup_location['lng']])

    if dropoff_location:
        folium.Marker(
            location=[dropoff_location['lat'], dropoff_location['lng']],
            popup=f"Dropoff: {row['dropoff_address']}",
            icon=folium.Icon(color='red')
        ).add_to(marker_group)
        # Add the dropoff location to the heatmap data
        heatmap_data.append([dropoff_location['lat'], dropoff_location['lng']])

# Add the heatmap to the heatmap group
HeatMap(heatmap_data).add_to(heatmap_group)

# Add the marker and heatmap groups to the map
marker_group.add_to(mymap)
heatmap_group.add_to(mymap)

# Add layer control to toggle the visibility of the layers
folium.LayerControl().add_to(mymap)

# Save the map to an HTML file
mymap.save("ride_map_with_heatmap_foreground.html")
```
